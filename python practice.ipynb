```python
#sum of even odd digit 
num = int(input("Enter number: "))

even_sum = 0
odd_sum = 0

while num > 0:
    digit = num % 10

    if digit % 2 == 0:
        even_sum += digit
    else:
        odd_sum += digit

    num //= 10

print("Sum of Even Digits =", even_sum)
print("Sum of Odd Digits =", odd_sum)
```

```python
#armstrong number
print("Armstrong Numbers:")

for num in range(1, 1001):
    temp = num
    total = 0

    while temp > 0:
        digit = temp % 10
        total += digit ** 3
        temp //= 10

    if total == num:
        print(num, end=" ")
```

```python
#hollow square star pattern
n = int(input("Enter size: "))

for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
#Count Uppercase, Lowercase, Digits and Spaces
text = input("Enter a string: ")

upper = lower = digit = space = 0

for ch in text:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1
    elif ch.isdigit():
        digit += 1
    elif ch == " ":
        space += 1

print("Uppercase:", upper)
print("Lowercase:", lower)
print("Digits:", digit)
print("Spaces:", space)
```

```python
#Frequency of Each Character
text = input("Enter a string: ")

freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

for key, value in freq.items():
    print(key, ":", value)
```

```python
#Print Unique Values in Original Order
numbers = []

print("Enter 10 numbers:")

for i in range(10):
    numbers.append(int(input()))

unique = []

for num in numbers:
    if num not in unique:
        unique.append(num)

print("Unique values:", unique)
```

```python
#Remove Duplicate Words While Preserving Order
sentence = input("Enter sentence: ")

words = sentence.split()

seen = set()

result = []

for word in words:
    if word not in seen:
        seen.add(word)
        result.append(word)

print(" ".join(result))
```

```python
#Student Marks and Topper
students = {
    "Aman": 85,
    "Rahul": 92,
    "Priya": 95,
    "Neha": 88
}

topper = max(students, key=students.get)

print("Topper:", topper)
print("Marks:", students[topper])
```

```python
#factorial using function
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact *= i

    return fact

num = int(input("Enter number: "))

print("Factorial =", factorial(num))
```

```python
def simple_interest(p, r, t):
    si = (p * r * t) / 100
    return si

def display(result):
    print("Simple Interest =", result)

p = float(input("Enter Principal: "))
r = float(input("Enter Rate: "))
t = float(input("Enter Time: "))

ans = simple_interest(p, r, t)
display(ans)
```
